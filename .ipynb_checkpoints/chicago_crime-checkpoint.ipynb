```python
#Declare dependencies
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
#Read in csv file
csv_path = "Resources/Crimes_-_2001_to_present (2010).csv"
chicago_crime_df = pd.read_csv(csv_path)
chicago_crime_df.head(5)
```

```python
#Trim dataset to include only caseNumber, date, and description
chicago_crime = chicago_crime_df[["Case Number", "Date", "Description"]]
chicago_crime.head(5)
```

```python
#Rename Description col to Crimes
chicago_crime.rename(index=str, columns={"Description":"Crimes"}, inplace=True)
chicago_crime.head(5)
```

```python
#clean up date to just year and place in new column Year
chicago_crime["Year"] = chicago_crime["Date"].str[6:10]
chicago_crime.head(5)
```

```python
#Group dataframe by year and count crimes occured
crime_year_totals = chicago_crime.groupby("Year").count()
crime_year_totals
```

```python
#Trim dataframe to only year and total crimes
ch_annual_crime = crime_year_totals[["Crimes"]]
ch_annual_crime
```

```python
#Convert crime totals to per capita values.
#Chicago has a population of 2.7 Million (2,700,000)
#A per 100,000 value would require dividing all totals by 27
ch_annual_crime["Per Capita (100K)"] = ch_annual_crime["Crimes"] / 27
ch_annual_crime
```

```python
#export final dataframe to csv
ch_annual_crime.to_csv("Resources/Chicago_Annual_Crime.csv", sep=",", encoding="utf-8")
```
