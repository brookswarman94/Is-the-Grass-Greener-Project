```python
# crime data SEATTLE--read file

seattle_df = pd.read_csv("crime_data_seattle.csv")
seattle_df.head()
```

```python
#SEATTLE
#select columns from data base to extract necessary columns
new_seattle_df = seattle_df[['Occurred Date', 'Crime Subcategory']]
#new_seattle_df.head()

new2_seattle_df = seattle_df[['Occurred Date', 'Crime Subcategory']]
new_seattle_df["Year"]=new_seattle_df["Occurred Date"].str[-4:]
new_seattle_df.head()

#new_seattle_df.describe()

new_seattle_df.groupby(['Year']).count().reset_index()
```

```python
#*******SEATTLE*******Select last 4 years = new_seattle4years_df

seattle4years_df = new_seattle_df.groupby(['Year']).count().reset_index()
new_seattle4years_df = seattle4years_df.iloc[[40,41,42,43],[0,1]]
new_seattle4years_df

#new_seattle4years_df = seattle4years_df.iloc[1:40,43]
#new_seattle4years_df

#data.iloc[[0,3,6,24], [0,5,6]] # 1st, 4th, 7th, 25th row + 1st 6th 7th columns.
#df_SN7577.iloc[1:4]
```

```python
#***********PLot Info SEATTLE-need 4 years**********
new_seattle4years_df.rename(columns={'Occurred Date':'Total Crimes S'}, inplace=True)
new_seattle4years_df

```

```python
#****************PHOENIX DATA********************
```

```python
phoenix_df = pd.read_csv("crime-data_phoenix.csv")
#phoenix_df.head()

new_phoenix_df = phoenix_df[['OCCURRED ON','UCR CRIME CATEGORY']]
new_phoenix_df["Year"]=new_phoenix_df["OCCURRED ON"].str[6:10]
#new_phoenix_df.head()

phoenix4years_df = new_phoenix_df.groupby(['Year']).count().reset_index()
phoenix4years_df
```

```python
#***********PLot Info SEATTLE-need 4 years**********
new_phoenix4years_df = phoenix4years_df.iloc[[0,1,2,3],[0,1]]
new_phoenix4years_df.rename(columns={'OCCURRED ON':'Total Crimes P'}, inplace=True)
new_phoenix4years_df
```

```python
#******DATA__ for Seatte Total CRIME BY CITY*****
# Merge file
Cities_Crime_df = pd.merge(new_seattle4years_df, new_phoenix4years_df)
Cities_Crime_df
```

```python
#*******Per Capita Calculations for Seattle and Phoenix**************

pop_units=100000
pop_seattle =704358/pop_units
pop_phoenix =1620000/pop_units
pop_chicago =2700000/pop_units

Cities_Crime_df["Phoenix Per Capita"] = Cities_Crime_df["Total Crimes P"]/pop_phoenix
Cities_Crime_df["Seattle Per Capita"] = Cities_Crime_df["Total Crimes S"]/pop_seattle
Cities_Crime_df
```

```python
#***********CHICAGO************************
```

```python
#import Chicago csv
chicago_df=pd.read_csv('Chicago_Annual_Crime.csv')
chicago_df

#join dataframe
final_df = pd.concat([Cities_Crime_df,chicago_df], axis=1)
final_df
```

```python
#Select desired colummns
final_df = final_df[["Year", "Phoenix Per Capita", "Seattle Per Capita", "Per Capita (100K)"]]
final_df
```

```python
#Rename Chicago "Per Capita" info
final_df.rename(index=str, columns={"Per Capita (100K)":"Chicago Per Capita"}, inplace=True)
final_df
```
